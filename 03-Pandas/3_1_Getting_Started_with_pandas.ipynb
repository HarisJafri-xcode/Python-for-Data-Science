# Sorting Rows

Rows can be sorted in ascending or descending order by the use of .sort_values() method.

```python
df_homeless.sort_values("individuals").head()
```

```python
df_homeless.sort_values("family_members", ascending=False).head()
```

```python
df_homeless.sort_values(['region','family_members'], ascending=[True,False]).head()
```

```python
df_homeless.sort_values(['region','family_members'], ascending=[False,True]).head()
```
